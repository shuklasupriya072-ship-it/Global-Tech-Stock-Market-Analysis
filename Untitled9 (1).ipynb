```python
import glob

# Load all 98 company files and merge
all_files = glob.glob('/content/stock_data/Technology Companies/*.csv')

df_list = []
for file in all_files:
    company = os.path.basename(file).replace('.csv', '')
    df = pd.read_csv(file)
    df['Company'] = company
    df_list.append(df)

# Merge all into one big dataframe
df_all = pd.concat(df_list, ignore_index=True)
print(f"Total rows: {df_all.shape[0]}")
print(f"Total columns: {df_all.shape[1]}")
print(f"Companies: {df_all['Company'].nunique()}")
```

```python
# Basic dataset overview
print("Date Range:")
df_all['Date'] = pd.to_datetime(df_all['Date'])
print(f"From: {df_all['Date'].min()}")
print(f"To: {df_all['Date'].max()}")

print(f"\nMissing Values:")
print(df_all.isnull().sum())

print(f"\nData Types:")
print(df_all.dtypes)
```

```python
# COVID period analysis - 2020 crash
covid_df = df_all[df_all['Date'].dt.year == 2020]

print(f"2020 data rows: {len(covid_df)}")
print(f"\n2020 Date range:")
print(f"From: {covid_df['Date'].min()}")
print(f"To: {covid_df['Date'].max()}")
print(f"\nCompanies in 2020: {covid_df['Company'].nunique()}")
```

```python
# Data Quality Check
print("=== DATA QUALITY REPORT ===")

# 1. Missing values
print(f"\n1. Missing Values:")
print(df_all.isnull().sum())

# 2. Duplicate rows
print(f"\n2. Duplicate Rows: {df_all.duplicated().sum()}")

# 3. Negative values check
print(f"\n3. Negative Price Values:")
print(f"   Close < 0: {(df_all['Close'] < 0).sum()}")
print(f"   Volume < 0: {(df_all['Volume'] < 0).sum()}")

# 4. Basic stats
print(f"\n4. Basic Statistics:")
print(df_all[['Open', 'High', 'Low', 'Close', 'Volume']].describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# COVID period - Jan to Nov 2020
covid = df_all[df_all['Date'].dt.year == 2020].copy()

# Each company ka % change 2020 mein
covid_summary = covid.groupby('Company').agg(
    Start_Price=('Close', 'first'),
    End_Price=('Close', 'last'),
    Min_Price=('Close', 'min')
).reset_index()

covid_summary['Annual_Return_%'] = ((covid_summary['End_Price'] - covid_summary['Start_Price']) / covid_summary['Start_Price'] * 100).round(2)
covid_summary['Max_Crash_%'] = ((covid_summary['Min_Price'] - covid_summary['Start_Price']) / covid_summary['Start_Price'] * 100).round(2)

# Top 10 best performers
top10 = covid_summary.nlargest(10, 'Annual_Return_%')
print(top10[['Company', 'Annual_Return_%', 'Max_Crash_%']])
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x='Company', y='Annual_Return_%', palette='viridis')
plt.title('Top 10 Tech Companies - COVID Recovery 2020', fontsize=16)
plt.xlabel('Company')
plt.ylabel('Annual Return %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Bottom 10 - Worst performers in 2020
bottom10 = covid_summary.nsmallest(10, 'Annual_Return_%')

plt.figure(figsize=(12, 6))
sns.barplot(data=bottom10, x='Company', y='Annual_Return_%', palette='Reds_r', hue='Company', legend=False)
plt.title('Top 10 Worst Performing Tech Companies - COVID 2020', fontsize=16)
plt.xlabel('Company')
plt.ylabel('Annual Return %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(bottom10[['Company', 'Annual_Return_%', 'Max_Crash_%']])
```

```python
# Year-wise average return for all companies
df_all['Year'] = df_all['Date'].dt.year

yearly = df_all.groupby(['Company', 'Year'])['Close'].agg(['first', 'last']).reset_index()
yearly['Annual_Return_%'] = ((yearly['last'] - yearly['first']) / yearly['first'] * 100).round(2)

# Overall average by year
avg_by_year = yearly.groupby('Year')['Annual_Return_%'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=avg_by_year, x='Year', y='Annual_Return_%', marker='o', color='steelblue', linewidth=2.5)
plt.axhline(y=0, color='red', linestyle='--', alpha=0.7)
plt.axvline(x=2020, color='orange', linestyle='--', alpha=0.7, label='COVID Year')
plt.title('Average Annual Return of Tech Companies (2006-2020)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Average Return %')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Q4 - Trading Volume Trends by Year
volume_by_year = df_all.groupby('Year')['Volume'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=volume_by_year, x='Year', y='Volume', hue='Year', palette='Blues_d', legend=False)
plt.title('Average Trading Volume by Year (2006-2020)', fontsize=16)
plt.xlabel('Year')
plt.ylabel('Average Volume')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q5 - Best performing companies overall (2006-2020)
overall = df_all.groupby('Company')['Close'].agg(['first', 'last']).reset_index()
overall['Total_Return_%'] = ((overall['last'] - overall['first']) / overall['first'] * 100).round(2)
top15_overall = overall.nlargest(15, 'Total_Return_%')

plt.figure(figsize=(14, 7))
sns.barplot(data=top15_overall, x='Company', y='Total_Return_%', hue='Company', palette='magma', legend=False)
plt.title('Top 15 Tech Companies - Overall Return 2006 to 2020', fontsize=16)
plt.xlabel('Company')
plt.ylabel('Total Return %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(top15_overall[['Company', 'Total_Return_%']])
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle('Global Tech Stock Market Analysis (2006-2020)', fontsize=20, fontweight='bold')

# Chart 1 - COVID Winners
sns.barplot(data=top10, x='Company', y='Annual_Return_%',
            hue='Company', palette='viridis', legend=False, ax=axes[0,0])
axes[0,0].set_title('Top 10 COVID Recovery Winners 2020', fontsize=13)
axes[0,0].set_xlabel('Company')
axes[0,0].set_ylabel('Annual Return %')
axes[0,0].tick_params(axis='x', rotation=45)

# Chart 2 - COVID Losers
sns.barplot(data=bottom10, x='Company', y='Annual_Return_%',
            hue='Company', palette='Reds_r', legend=False, ax=axes[0,1])
axes[0,1].set_title('Top 10 COVID Worst Performers 2020', fontsize=13)
axes[0,1].set_xlabel('Company')
axes[0,1].set_ylabel('Annual Return %')
axes[0,1].tick_params(axis='x', rotation=45)

# Chart 3 - Year wise trend
sns.lineplot(data=avg_by_year, x='Year', y='Annual_Return_%',
             marker='o', color='steelblue', linewidth=2.5, ax=axes[1,0])
axes[1,0].axhline(y=0, color='red', linestyle='--', alpha=0.7)
axes[1,0].set_title('Average Annual Return Trend (2006-2020)', fontsize=13)
axes[1,0].set_xlabel('Year')
axes[1,0].set_ylabel('Average Return %')

# Chart 4 - Overall best performers
sns.barplot(data=top15_overall.head(10), x='Company', y='Total_Return_%',
            hue='Company', palette='magma', legend=False, ax=axes[1,1])
axes[1,1].set_title('Top 10 Overall Performers (2006-2020)', fontsize=13)
axes[1,1].set_xlabel('Company')
axes[1,1].set_ylabel('Total Return %')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('stock_analysis_summary.png', dpi=150, bbox_inches='tight')
```

```python
# Moving Average - AAPL example
aapl = df_all[df_all['Company'] == 'AAPL'].copy()
aapl = aapl.sort_values('Date')
aapl['MA_50'] = aapl['Close'].rolling(window=50).mean()
aapl['MA_200'] = aapl['Close'].rolling(window=200).mean()

plt.figure(figsize=(14, 6))
plt.plot(aapl['Date'], aapl['Close'], label='Close Price', alpha=0.5)
plt.plot(aapl['Date'], aapl['MA_50'], label='50 Day MA', color='orange')
plt.plot(aapl['Date'], aapl['MA_200'], label='200 Day MA', color='red')
plt.title('Apple Stock - Moving Average Analysis', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Volatility - which company most risky
df_all['Daily_Return'] = df_all.groupby('Company')['Close'].pct_change()
volatility = df_all.groupby('Company')['Daily_Return'].std().reset_index()
volatility.columns = ['Company', 'Volatility']
volatility = volatility.nlargest(10, 'Volatility')

plt.figure(figsize=(12, 6))
sns.barplot(data=volatility, x='Company', y='Volatility',
            hue='Company', palette='rocket', legend=False)
plt.title('Top 10 Most Volatile Tech Companies', fontsize=16)
plt.xlabel('Company')
plt.ylabel('Volatility (Std Dev of Daily Returns)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Which companies move together
pivot = df_all.pivot_table(index='Date', columns='Company', values='Close')
correlation = pivot.corr()

# Top 10 companies only
top_companies = ['AAPL', 'NVDA', 'MSFT', 'CRM', 'SHOP', 'NET', 'CSCO', 'IBM', 'INTC', 'SAP']
corr_subset = correlation.loc[top_companies, top_companies]

plt.figure(figsize=(12, 10))
sns.heatmap(corr_subset, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Stock Price Correlation - Top Tech Companies', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import sqlite3
import pandas as pd

# Create database
conn = sqlite3.connect('stock_analysis.db')

# Save data to SQL table
df_all.to_sql('tech_stocks', conn, if_exists='replace', index=False)

print("Database created successfully!")
print(f"Total rows: {pd.read_sql('SELECT COUNT(*) as total FROM tech_stocks', conn).iloc[0,0]}")
```

```python
# SQL Query 1 - Top 10 best performing companies overall
query1 = """
SELECT
    Company,
    ROUND(MIN(Close), 2) as Min_Price,
    ROUND(MAX(Close), 2) as Max_Price,
    ROUND(((MAX(Close) - MIN(Close)) / MIN(Close)) * 100, 2) as Total_Return_Pct
FROM tech_stocks
GROUP BY Company
ORDER BY Total_Return_Pct DESC
LIMIT 10;
"""

result1 = pd.read_sql(query1, conn)
print("Top 10 Best Performing Companies (Overall):")
print(result1)
```

```python
# SQL Query 2 - Average yearly return by year
query2 = """
SELECT
    strftime('%Y', Date) as Year,
    ROUND(AVG(Close), 2) as Avg_Close_Price,
    ROUND(MAX(Close), 2) as Max_Price,
    ROUND(MIN(Close), 2) as Min_Price
FROM tech_stocks
GROUP BY Year
ORDER BY Year;
"""

result2 = pd.read_sql(query2, conn)
print("Yearly Average Stock Price:")
print(result2)
```

```python
# SQL Query 3 - Most volatile companies
query3 = """
SELECT
    Company,
    ROUND(AVG(High - Low), 2) as Avg_Daily_Range,
    ROUND(MAX(High - Low), 2) as Max_Daily_Range
FROM tech_stocks
GROUP BY Company
ORDER BY Avg_Daily_Range DESC
LIMIT 10;
"""
result3 = pd.read_sql(query3, conn)
print("Query 3 - Most Volatile Companies:")
print(result3)
```

```python
# SQL Query 4 - Best year for each company
query4 = """
SELECT Company, Year, MAX(Annual_Avg) as Best_Year_Avg
FROM (
    SELECT Company,
        strftime('%Y', Date) as Year,
        ROUND(AVG(Close), 2) as Annual_Avg
    FROM tech_stocks
    GROUP BY Company, Year
)
GROUP BY Company
ORDER BY Best_Year_Avg DESC
LIMIT 10;
"""
result4 = pd.read_sql(query4, conn)
print("Query 4 - Best Year Average for Each Company:")
print(result4)
```

```python
# SQL Query 5 - Companies above average price
query5 = """
SELECT Company,
    ROUND(AVG(Close), 2) as Avg_Price
FROM tech_stocks
GROUP BY Company
HAVING AVG(Close) > (SELECT AVG(Close) FROM tech_stocks)
ORDER BY Avg_Price DESC
LIMIT 10;
"""
result5 = pd.read_sql(query5, conn)
print("Query 5 - Companies Above Average Price:")
print(result5)

```

```python
# Bonus - CTE Query
query_cte = """
WITH yearly_returns AS (
    SELECT
        Company,
        strftime('%Y', Date) as Year,
        ROUND(AVG(Close), 2) as Avg_Close
    FROM tech_stocks
    GROUP BY Company, Year
),
company_growth AS (
    SELECT
        Company,
        MIN(Avg_Close) as Min_Avg,
        MAX(Avg_Close) as Max_Avg,
        ROUND((MAX(Avg_Close) - MIN(Avg_Close)) / MIN(Avg_Close) * 100, 2) as Growth_Pct
    FROM yearly_returns
    GROUP BY Company
)
SELECT * FROM company_growth
ORDER BY Growth_Pct DESC
LIMIT 10;
"""
result_cte = pd.read_sql(query_cte, conn)
print("CTE Query - Company Growth Analysis:")
print(result_cte)
```
